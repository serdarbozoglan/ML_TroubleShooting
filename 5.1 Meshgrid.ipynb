```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def cos2d(x,y):
    return np.cos(x) + np.cos(y)
```

`x => 0..4 y => 0..4`

```python
x[0,0] = 0
x[0,1] = 0
x[0,2] = 0
..

y[0,0] = 0
y[0,1] = 1

x[1,2] = 1 x'in aldigi degeri gosterir, x component is 1 
y[0,2] = 2 y=2 oldugu icin 1'dir, y component is 1
```

```python
xvalues = np.array([0,1,2,3,4])
yvalues = np.array([0,1,2,3,4])
```

```python
xx, yy = np.meshgrid(xvalues, yvalues)

plt.plot(xx, yy, marker='.', color='b', linestyle='none')
plt.show()

# we're creating a data points for example while x=1 for all y values, it should map to x=1 line
```

```python
xx, yy = np.meshgrid(np.linspace(0,2 * np.pi, 100), np.linspace(0,2 * np.pi, 100))
```

```python
zz = cos2d(xx ,yy)
```

```python
plt.imshow(zz, origin='lower', interpolation='none')
plt.show()
```
