```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.head()
```

#### BEFORE PCA NORMALIZE THE DATA

```python
df_norm = (df-df.mean())/df.std()
```

```python
pca = PCA(n_components=4)
pca.fit_transform(df_norm.values)
```

```python
pca.explained_variance_ratio_
# we might say 3 component is enough to explain variation on y
```

```python
df.shape
```

```python
np.identity(df.shape[1])
```

```python
components = pca.transform(np.identity(df.shape[1]))
components
```

```python
pd.DataFrame(components, columns=['pc1', 'pc2', 'pc3', 'pc4'], index=df.columns)
```

```python
## for pricipal component1, speal length (%51), petal length(58) and petal with (56%) explain the variation on y through pca
```
